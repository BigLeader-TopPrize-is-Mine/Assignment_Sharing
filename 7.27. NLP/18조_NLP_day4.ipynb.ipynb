```python
# 소문자 변환 진행
def pre_func(title):
    res=''.join(w.lower() for w in title if w not in punctuation)
    return res

pre_headline=[pre_func(x) for x in headline]
pre_headline[0:30]
```

1. 불용어 제거 사전에 불용어 추가

```python
#### 전처리부분 (import nltk / from nltk.corpus import stopwords -> nltk.download('stopwords') 필요)
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

post_headline = []

for sentence in pre_headline :
    post_headline.append(' '.join([word for word in sentence.split(' ') if word not in stopwords.words('english')]))

print(post_headline)
```

```python
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Embedding, SimpleRNN, LSTM
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

tok=Tokenizer()
tok.fit_on_texts(post_headline)
tok.index_word
len(tok.index_word)
```

```python
vocab_size=len(tok.index_word)+1 # 0 부터시작하기 때문에, 1을 더 함
```

```python
sequences=[]

for s in post_headline:
    # enc=print(tok.texts_to_sequences([s])[0]) # 각 문장별 인코딩
    enc=tok.texts_to_sequences([s])[0]

    for i in range(1, len(enc)):
        seq=enc[:i+1]
        sequences.append(seq)
```

```python
ml=max(len(i) for i in sequences)
print(ml) # 가장 긴 문장은 24개의 단어를 가진다.
```

```python
sequences_matrix=pad_sequences(sequences, maxlen=ml)
```

```python
sequences_x=sequences_matrix[:,:-1]
sequences_y=sequences_matrix[:,-1]
y=to_categorical(sequences_y, vocab_size)

print(sequences_matrix.shape)
print(sequences_x.shape)
print(y.shape)
print(sequences_y)
```

2. 수업 때 한 신경망 모델 구조 변경 및 성능 개선

```python
model=Sequential()

# vocab_size
model.add(Embedding(vocab_size, 100)) # 3620 차원 → 10 차원으로 임베딩
model.add(LSTM(128, return_sequences=True))
model.add(LSTM(64, return_sequences=False))
model.add(Dense(vocab_size, activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy',
              optimizer='adam',metrics=['accuracy'])
model.fit(sequences_x, y, epochs=200, verbose=2)
```

```python
def gen_sent(model, tok, c_word, n):
    pred_sent=''

    # 구현 할 부분
    for _ in range(n):
        enc=tok.texts_to_sequences([c_word])[0]
        enc=pad_sequences([enc], maxlen=ml-1, padding='pre')
        # print(enc)
        res=model.predict(enc)
        res=np.argmax(res)

        for w, i in tok.word_index.items():
            if i==res:
                break
        # print("예측 단어 :", w)
        c_word=c_word+' '+ w            # 입력 단어에 다음 단어 추가

    print(c_word)
```
