```python
import pandas as pd
import numpy as np

df = pd.read_csv("./data/tumor.csv")
df.describe()
```

```python
# for i in df['diagnosis'].unique() :
#     print(df['diagnosis'].count())

print(df["diagnosis"].groupby(df["diagnosis"]).count())
df["diagnosis"].unique()
# print(df.columns)

feature = ['radius_mean', 'texture_mean', 'perimeter_mean',
       'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave.points_mean', 'symmetry_mean', 'fractal_dimension_mean']

X = df[feature]
y = df['diagnosis']

print(np.corrcoef(X))

#회귀분석 위해 y 추가
df["diagnosis_category"] = [0 if i == 'B' else 1 for i in df['diagnosis']]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=1004)
```

```python
from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()
std_scaler.fit(X_train)

X_train_std = std_scaler.transform(X_train)
X_test_std = std_scaler.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
clf_knn = KNeighborsClassifier(n_neighbors=4)
clf_knn.fit(X_train_std, y_train)
```

```python
knn_pred = clf_knn.predict(X_test_std)

from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,knn_pred))

from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test,knn_pred))

from sklearn.metrics import classification_report
print(classification_report(y_test,knn_pred))
```

```python
# y범주형으로 바꿈
y = df["diagnosis_category"]
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=1004)

std_scaler = StandardScaler()
std_scaler.fit(X_train)
X_train_std = std_scaler.transform(X_train)
X_test_std = std_scaler.transform(X_test)

from sklearn.linear_model import LinearRegression

clf_lr = LinearRegression()
clf_lr.fit(X_train_std, y_train)

pred_lr = clf_lr.predict(X_test_std)

from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_test,pred_lr))
```

```python
#lasso, ridge, elasticnet
from sklearn.linear_model import Lasso

clf_lasso = Lasso(alpha=0.01)
clf_lasso.fit(X_train_std,y_train)
pred_lasso = clf_lasso.predict(X_test_std)

print(mean_squared_error(y_test,pred_lasso))

from sklearn.linear_model import Ridge

clf_ridge = Ridge()
clf_ridge.fit(X_train_std,y_train)
pred_ridge = clf_ridge.predict(X_test_std)

print(mean_squared_error(y_test,pred_ridge))

from sklearn.linear_model import ElasticNet

clf_elas = ElasticNet(alpha=0.01,l1_ratio=0.01)
clf_elas.fit(X_train_std,y_train)
pred_elas = clf_elas.predict(X_test_std)

print(mean_squared_error(y_test,pred_elas))
```

```python
#Logistic
y = df['diagnosis']
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=1004)

std_scaler = StandardScaler()
std_scaler.fit(X_train)

X_train_std = std_scaler.transform(X_train)
X_test_std = std_scaler.transform(X_test)

from sklearn.linear_model import LogisticRegression

clf_logi = LogisticRegression(penalty='l2')
clf_logi.fit(X_train_std,y_train)

pred_logi = clf_logi.predict(X_test_std)
pred_logi_2 = clf_logi.predict_proba(X_test_std)
print(pred_logi_2) #ㅇㄱㅇ

from sklearn.metrics import recall_score
print(recall_score(y_test,pred_logi,average='macro'))
```

```python
from sklearn.naive_bayes import GaussianNB

y = df['diagnosis_category']
X_train, X_test, y_train, y_test = train_test_split(X,y,stratify=y,random_state=1004)

std_scaler = StandardScaler()
std_scaler.fit(X_train)

X_train_std = std_scaler.transform(X_train)
X_test_std = std_scaler.transform(X_test)


clf_nb = GaussianNB()
clf_nb.fit(X_train,y_train)

pred_nb = clf_nb.predict(X_test)

# accu_score = accuracy_score(y_test, pred_nb)
# print(accu_score)

recl_score = recall_score(y_test,pred_nb,average="binary")
print(recl_score)

# conf_matrix = confusion_matrix(y_test,pred_nb)
# print(conf_matrix)

# class_report = classification_report(y_test,pred_nb)
# print(class_report)
```

```python
from sklearn import tree

clf_tree = tree.DecisionTreeClassifier(criterion="entropy",random_state=1004)
clf_tree.fit(X_train_std,y_train)

pred_tree = clf_tree.predict(X_test_std)
# print(pred_tree)

# -------------------------------> 클래스 값이 이진이면 binary 쓰면 왜 높아짐?
recl_score = recall_score(y_test,pred_tree,average='binary')
print(recl_score)

# conf_matrix = confusion_matrix(y_test, pred_tree)
# print(conf_matrix)

class_report = classification_report(y_test,pred_tree)
print(class_report)
```
