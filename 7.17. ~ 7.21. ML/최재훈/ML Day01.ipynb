$$ \prod_{x=1}^10 x = 1 \times 2 \times 3 \times \cdots \times 10 $$

```python
res = 1
for i in range(1, 11):
    res *= i
    print(i, res)
print(res)
```

$$ n! = n \times (n-1) \times (n-2) \times \cdots \times 2 \times 1 $$

```python
def factorial(n):
    res = 1
    for i in range(1, n+1):
        res *= i
    return res

factorial(5)
```

$$ \begin{pmatrix} n \\ x \end{pmatrix} = nCx = \frac{n!}{x!(n-x)!}$$

```python
# 머신러닝에서 피쳐를 다 쓰는 방법이 있고 특정 피쳐만 뽑는 방식이 있는데, 후자의 경우 combinations이 사용된다.
def combinations(n, r):
    return (factorial(n))/(factorial(r)*factorial(n-r))

combinations(4, 2)
```

$$ min(x) \;\; argmin(x) $$

```python
x = [4,2,5,9,3]
print(min(x)) # -> 최솟값
print(x.index(min(x)))# argmin(x) -> 최솟값의 인덱스
```
